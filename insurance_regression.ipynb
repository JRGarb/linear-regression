```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

This program will predict how age affects health insurance costs using linear regression.

```python
# Dataframe is created.
df = pd.read_csv('insurance.csv')
df.head()
```

```python
# Plotting the data.
plt.figure()
sns.scatterplot(data=df, x=df.age, y=df.charges)
plt.xlabel('Age')
plt.ylabel('Insurance Charges ($)')
plt.show()
plt.close()
```

```python
# Fitting the model.
X = df['age'].to_numpy().reshape(-1, 1)
y = df['charges'].to_numpy()

insurance_model = LinearRegression()
insurance_model.fit(X, y)
```

```python
# Predicting based on age.
y_pred = insurance_model.predict(X)
```

```python
# Plotting the model and data.
plt.figure()
sns.scatterplot(data=df, x=df.age, y=df.charges)
plt.title('Insurance charges vs. Age')
plt.xlabel('Age')
plt.ylabel('Insurance Charges ()')
plt.locator_params(axis= 'x', nbins=20)
plt.plot(X, y_pred, color='r')
plt.show()
```
